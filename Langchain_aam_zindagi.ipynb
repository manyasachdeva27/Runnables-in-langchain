```python
import random

#1st component
class NakliLLM:
  def __init__(self):
    print('LLM created')

  def predict(self,prompt):
    response_list=[
        "Delhi is the capital of India.",
        "The Eiffel Tower is a famous landmark in Paris.",
        "The Mona Lisa is a famous painting by Leonardo da Vinci"
    ]
    return {'response':random.choice(response_list)}
```

```python
llm=NakliLLM()
```

```python
llm.predict('who created Mona Lisa')
```

```python
#2nd component
class NakliPromptTemplate:
  def __init__(self,template,input_variables):
    self.template=template
    self.input_variables=input_variables

  def format(self,input_dict):
    return self.template.format(**input_dict)
```

```python
template=NakliPromptTemplate(
    template='what is the {length} poem about {country}',
    input_variables=['length','country']
)
```

```python
prompt=template.format(input_dict={'country':'India','length':'5 lines'})
```

```python
llm=NakliLLM()
```

```python
llm.predict(prompt)
```

```python
class NakliLLMchain:
  def __init__(self,llm,prompt):
    self.llm=llm
    self.prompt=prompt

  def run(self,input_dict):
    final_prompt=self.prompt.format(input_dict)
    result=self.llm.predict(final_prompt)

    return result['response']
```

```python
template=NakliPromptTemplate(
    template='what is the {length} poem about {country}',
    input_variables=['length','country']
)
```

```python
llm=NakliLLM()
```

```python
chain=NakliLLMchain(llm,template)
```

```python
chain.run({'length':"5 lines",'country':'Paris'})
```
